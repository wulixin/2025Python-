```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
plt.rcParams['font.sans-serif'] = ['PingFang SC', 'Arial Unicode MS']  # Mac常用中文字体
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题
```

```python
def generate_mock_data_part2():
    """生成第9-13页所需的模拟数据"""
    data = {}

    # 图9: 固定资产投资
    dates_invest = pd.date_range(start='2015-02-01', end='2025-09-01', freq='M')
    invest_total = np.random.uniform(-5, 15, len(dates_invest))
    state_invest = np.random.uniform(-8, 20, len(dates_invest))
    private_invest = np.random.uniform(-15, 10, len(dates_invest))

    infra_invest = np.random.uniform(-10, 25, len(dates_invest))
    manu_invest = np.random.uniform(-5, 15, len(dates_invest))
    real_estate_invest = np.random.uniform(-20, 10, len(dates_invest))

    data['fixed_investment'] = {
        'dates': dates_invest,
        'invest_total': invest_total,
        'state_invest': state_invest,
        'private_invest': private_invest,
        'infra_invest': infra_invest,
        'manu_invest': manu_invest,
        'real_estate_invest': real_estate_invest,
    }

    # 图10: 工业生产
    dates_industry = pd.date_range(start='2015-02-01', end='2025-09-01', freq='M')
    industrial_value_added = np.random.uniform(-10, 15, len(dates_industry))
    industrial_profit = np.random.uniform(-20, 30, len(dates_industry))

    # PMI分项雷达图数据 (三个时间点)
    pmi_items = ['生产', '新订单', '从业人员', '供应商配送时间', '原材料库存', '产成品库存', '采购量', '进口', '出厂价格', '原材料购进价格']
    pmi_2412 = np.random.uniform(45, 55, len(pmi_items))
    pmi_2506 = np.random.uniform(44, 54, len(pmi_items))
    pmi_2509 = np.random.uniform(43, 53, len(pmi_items))

    data['industrial_production'] = {
        'dates': dates_industry,
        'industrial_value_added': industrial_value_added,
        'industrial_profit': industrial_profit,
        'pmi_items': pmi_items,
        'pmi_2412': pmi_2412,
        'pmi_2506': pmi_2506,
        'pmi_2509': pmi_2509,
    }

    # 图11: 就业和收入
    dates_unemployment = pd.date_range(start='2018-01-01', end='2025-09-01', freq='M')
    unemployment_overall = np.random.uniform(4, 6, len(dates_unemployment))
    unemployment_16_24 = np.random.uniform(12, 19, len(dates_unemployment))
    unemployment_25_29 = np.random.uniform(5, 7, len(dates_unemployment))
    unemployment_30_59 = np.random.uniform(3, 4, len(dates_unemployment))

    dates_income = pd.date_range(start='2018-01-01', end='2025-09-01', freq='M')
    per_capita_income = np.random.uniform(-5, 15, len(dates_income))
    income_expectation_index = np.random.uniform(85, 125, len(dates_income))

    data['employment_income'] = {
        'dates_unemp': dates_unemployment,
        'unemployment_overall': unemployment_overall,
        'unemployment_16_24': unemployment_16_24,
        'unemployment_25_29': unemployment_25_29,
        'unemployment_30_59': unemployment_30_59,
        'dates_income': dates_income,
        'per_capita_income': per_capita_income,
        'income_expectation_index': income_expectation_index,
    }

    # 图12: 通货膨胀和实际利率
    dates_cpi = pd.date_range(start='2014-01-01', end='2025-09-01', freq='M')
    cpi = np.random.uniform(-1, 6, len(dates_cpi))
    ppi = np.random.uniform(-6, 19, len(dates_cpi))

    dates_rate = pd.date_range(start='2014-01-01', end='2025-09-01', freq='M')
    real_deposit_rate = np.random.uniform(-10, 12, len(dates_rate))
    real_loan_rate = np.random.uniform(-10, 15, len(dates_rate))

    data['inflation_rates'] = {
        'dates_cpi': dates_cpi,
        'cpi': cpi,
        'ppi': ppi,
        'dates_rate': dates_rate,
        'real_deposit_rate': real_deposit_rate,
        'real_loan_rate': real_loan_rate,
    }

    # 图13: 对外贸易
    dates_trade = pd.date_range(start='2015-01-01', end='2025-09-01', freq='M')
    export_growth = np.random.uniform(-20, 50, len(dates_trade))
    import_growth = np.random.uniform(-20, 50, len(dates_trade))
    goods_trade_balance = np.random.uniform(-400, 1400, len(dates_trade))
    services_trade_balance = np.random.uniform(-400, 200, len(dates_trade))

    # 主要出口市场占比 (条形图)
    markets = ['东盟', '欧盟', '美国', '非洲', '日本', '印度', '俄罗斯']
    shares_2010 = np.random.uniform(5, 20, len(markets))
    shares_2017 = np.random.uniform(5, 20, len(markets))
    shares_2024 = np.random.uniform(5, 20, len(markets))
    shares_2025 = np.random.uniform(5, 20, len(markets))

    data['foreign_trade'] = {
        'dates': dates_trade,
        'export_growth': export_growth,
        'import_growth': import_growth,
        'goods_trade_balance': goods_trade_balance,
        'services_trade_balance': services_trade_balance,
        'markets': markets,
        'shares_2010': shares_2010,
        'shares_2017': shares_2017,
        'shares_2024': shares_2024,
        'shares_2025': shares_2025,
    }

    return data

# 生成模拟数据
mock_data_part2 = generate_mock_data_part2()
```

## 图6: 中国经济：固定资产投资

```python
def plot_fixed_investment(data):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 7))

    # 左侧图：固定资产投资完成额
    ax1.plot(data['dates'], data['invest_total'], color='black', label='投资总额')
    ax1.plot(data['dates'], data['state_invest'], color='orange', label='国有及国有控股企业投资')
    ax1.plot(data['dates'], data['private_invest'], color='purple', label='民间投资')
    ax1.set_title('固定资产投资完成额\n累计同比', fontsize=14)
    ax1.set_ylabel('百分比 (%)')
    ax1.legend()

    # 右侧图：主要行业投资增速
    ax2.plot(data['dates'], data['infra_invest'], color='purple', label='基础设施建设投资')
    ax2.plot(data['dates'], data['manu_invest'], color='blue', label='制造业固定资产投资')
    ax2.plot(data['dates'], data['real_estate_invest'], color='black', label='房地产开发投资')
    ax2.set_title('主要行业投资增速\n累计同比', fontsize=14)
    ax2.set_ylabel('百分比 (%)')
    ax2.legend()

    plt.tight_layout()
    plt.show()

# 调用绘图函数
plot_fixed_investment(mock_data_part2['fixed_investment'])
```

## 图7: 中国经济：工业生产

```python
def plot_industrial_production(data):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 7))

    # 左侧图：工业增加值和工业企业利润
    ax1.plot(data['dates'], data['industrial_value_added'], color='blue', label='工业增加值')
    ax1.set_ylabel('同比 (%)', color='blue')
    ax1.tick_params(axis='y', labelcolor='blue')

    ax1b = ax1.twinx()
    ax1b.plot(data['dates'], data['industrial_profit'], color='orange', label='工业企业利润')
    ax1b.set_ylabel('累计同比* (%)', color='orange')
    ax1b.tick_params(axis='y', labelcolor='orange')

    ax1.set_title('工业增加值和工业企业利润*\n同比', fontsize=14)
    ax1.legend(loc='upper left')
    ax1b.legend(loc='upper right')

    # 右侧图：官方制造业PMI主要分项 (雷达图)
    categories = data['pmi_items']
    N = len(categories)

    # 计算角度
    angles = np.linspace(0, 2 * np.pi, N, endpoint=False).tolist()
    angles += angles[:1]  # 闭合图形

    fig2, ax2 = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

    # 绘制三条线
    values_2412 = data['pmi_2412'].tolist()
    values_2412 += values_2412[:1]
    ax2.plot(angles, values_2412, 'o-', linewidth=2, label='24年12月')
    ax2.fill(angles, values_2412, alpha=0.25)

    values_2506 = data['pmi_2506'].tolist()
    values_2506 += values_2506[:1]
    ax2.plot(angles, values_2506, 'o-', linewidth=2, label='25年6月')
    ax2.fill(angles, values_2506, alpha=0.25)

    values_2509 = data['pmi_2509'].tolist()
    values_2509 += values_2509[:1]
    ax2.plot(angles, values_2509, 'o-', linewidth=2, label='25年9月')
    ax2.fill(angles, values_2509, alpha=0.25)

    # 设置标签和标题
    ax2.set_thetagrids(np.degrees(angles[:-1]), categories)
    ax2.set_title('官方制造业PMI**主要分项\n指数', size=14)
    ax2.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

    plt.tight_layout()
    plt.show()

# 调用绘图函数
plot_industrial_production(mock_data_part2['industrial_production'])
```

## 图8: 中国经济：就业和收入

```python
def plot_employment_income(data):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 7))

    # 左侧图：城镇调查失业率
    ax1.plot(data['dates_unemp'], data['unemployment_overall'], color='purple', label='总体')
    ax1.plot(data['dates_unemp'], data['unemployment_16_24'], color='black', label='16 - 24岁 (不含在校生)')
    ax1.plot(data['dates_unemp'], data['unemployment_25_29'], color='blue', label='25 - 29岁 (不含在校生)')
    ax1.plot(data['dates_unemp'], data['unemployment_30_59'], color='red', label='30 - 59岁 (不含在校生)')
    ax1.set_title('城镇调查失业率*', fontsize=14)
    ax1.set_ylabel('百分比 (%)')
    ax1.legend()

    # 右侧图：居民收入和预期
    ax2.plot(data['dates_income'], data['per_capita_income'], color='black', label='人均可支配收入')
    ax2.set_ylabel('实际累计同比 (%)', color='black')
    ax2.tick_params(axis='y', labelcolor='black')

    ax2b = ax2.twinx()
    ax2b.plot(data['dates_income'], data['income_expectation_index'], color='blue', label='消费者收入预期指数 (右轴)')
    ax2b.set_ylabel('指数', color='blue')
    ax2b.tick_params(axis='y', labelcolor='blue')

    ax2.set_title('居民收入和预期\n实际累计同比', fontsize=14)
    ax2.legend(loc='upper left')
    ax2b.legend(loc='upper right')

    plt.tight_layout()
    plt.show()

# 调用绘图函数
plot_employment_income(mock_data_part2['employment_income'])
```

## 图9: 中国经济：通货膨胀和实际利率

```python
def plot_inflation_rates(data):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 7))

    # 左侧图：CPI和PPI
    ax1.plot(data['dates_cpi'], data['cpi'], color='black', label='CPI')
    ax1.set_ylabel('同比 (%)', color='black')
    ax1.tick_params(axis='y', labelcolor='black')

    ax1b = ax1.twinx()
    ax1b.plot(data['dates_cpi'], data['ppi'], color='blue', label='PPI (右轴)')
    ax1b.set_ylabel('同比 (%)', color='blue')
    ax1b.tick_params(axis='y', labelcolor='blue')

    ax1.set_title('居民消费价格指数 (CPI) 和工业生产者出厂价格指数 (PPI)\n同比', fontsize=14)
    ax1.legend(loc='upper left')
    ax1b.legend(loc='upper right')

    # 右侧图：实际利率
    ax2.plot(data['dates_rate'], data['real_deposit_rate'], color='black', label='实际存款利率*')
    ax2.plot(data['dates_rate'], data['real_loan_rate'], color='blue', label='实际贷款利率**')
    ax2.set_title('实际利率', fontsize=14)
    ax2.set_ylabel('百分比 (%)')
    ax2.legend()

    plt.tight_layout()
    plt.show()

# 调用绘图函数
plot_inflation_rates(mock_data_part2['inflation_rates'])
```

## 图10  中国经济：对外贸易

```python
def plot_foreign_trade(data):
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    fig.suptitle('中国经济：对外贸易', fontsize=16)

    # 左上图：货物贸易总额
    ax1 = axes[0, 0]
    ax1.plot(data['dates'], data['export_growth'], color='black', label='出口')
    ax1.plot(data['dates'], data['import_growth'], color='purple', label='进口')
    ax1.set_title('货物贸易总额\n同比，美元计价，3个月滚动区间', fontsize=12)
    ax1.set_ylabel('百分比 (%)')
    ax1.legend()

    # 左下图：经常项目差额
    ax2 = axes[1, 0]
    ax2.plot(data['dates'], data['goods_trade_balance'], color='black', label='货物贸易')
    ax2.plot(data['dates'], data['services_trade_balance'], color='purple', label='服务贸易')
    ax2.set_title('经常项目差额\n亿美元，3个月移动平均', fontsize=12)
    ax2.set_ylabel('亿美元')
    ax2.legend()

    # 右侧图：主要出口市场
    ax3 = axes[0, 1]
    ax3.axis('off')  # 关闭坐标轴，因为我们只画条形图

    # 创建子图用于条形图
    ax4 = fig.add_subplot(2, 2, 4)
    y_pos = np.arange(len(data['markets']))
    bar_width = 0.2

    ax4.barh(y_pos, data['shares_2010'], bar_width, color='orange', label='2010年')
    ax4.barh(y_pos + bar_width, data['shares_2017'], bar_width, color='gray', label='2017年')
    ax4.barh(y_pos + 2 * bar_width, data['shares_2024'], bar_width, color='purple', label='2024年')
    ax4.barh(y_pos + 3 * bar_width, data['shares_2025'], bar_width, color='blue', label='2025年1-8月')

    ax4.set_yticks(y_pos + 1.5 * bar_width)
    ax4.set_yticklabels(data['markets'])
    ax4.set_xlabel('出口总额占比 (%)')
    ax4.set_title('主要出口市场\n出口总额占比', fontsize=12)
    ax4.legend()

    plt.tight_layout()
    plt.show()

# 调用绘图函数
plot_foreign_trade(mock_data_part2['foreign_trade'])
```

## 第三步：构建指标评估体系

```python
def evaluate_indicators_part2(data):
    """评估第9-13页的关键经济指标"""
    evaluation = {}

    # 9. 固定资产投资
    latest_invest_total = data['fixed_investment']['invest_total'][-1]
    if latest_invest_total > 5:
        evaluation['fixed_investment'] = '健康'
    elif latest_invest_total > 0:
        evaluation['fixed_investment'] = '关注'
    else:
        evaluation['fixed_investment'] = '风险'

    # 10. 工业生产
    latest_industrial_value_added = data['industrial_production']['industrial_value_added'][-1]
    latest_industrial_profit = data['industrial_production']['industrial_profit'][-1]
    if latest_industrial_value_added > 5 and latest_industrial_profit > 5:
        evaluation['industrial_production'] = '健康'
    elif latest_industrial_value_added > 0 or latest_industrial_profit > 0:
        evaluation['industrial_production'] = '关注'
    else:
        evaluation['industrial_production'] = '风险'

    # 11. 就业和收入
    latest_unemployment_overall = data['employment_income']['unemployment_overall'][-1]
    latest_per_capita_income = data['employment_income']['per_capita_income'][-1]
    if latest_unemployment_overall < 5 and latest_per_capita_income > 5:
        evaluation['employment_income'] = '健康'
    elif latest_unemployment_overall < 6 or latest_per_capita_income > 0:
        evaluation['employment_income'] = '关注'
    else:
        evaluation['employment_income'] = '风险'

    # 12. 通货膨胀和实际利率
    latest_cpi = data['inflation_rates']['cpi'][-1]
    latest_ppi = data['inflation_rates']['ppi'][-1]
    latest_real_loan_rate = data['inflation_rates']['real_loan_rate'][-1]
    if 1 <= latest_cpi <= 3 and latest_ppi > -2 and latest_real_loan_rate > 0:
        evaluation['inflation_rates'] = '健康'
    elif 0 <= latest_cpi <= 4 and latest_ppi > -5 and latest_real_loan_rate > -2:
        evaluation['inflation_rates'] = '关注'
    else:
        evaluation['inflation_rates'] = '风险'

    # 13. 对外贸易
    latest_export_growth = data['foreign_trade']['export_growth'][-1]
    latest_import_growth = data['foreign_trade']['import_growth'][-1]
    if latest_export_growth > 5 and latest_import_growth > 5:
        evaluation['foreign_trade'] = '健康'
    elif latest_export_growth > 0 or latest_import_growth > 0:
        evaluation['foreign_trade'] = '关注'
    else:
        evaluation['foreign_trade'] = '风险'

    return evaluation

# 执行评估
eval_results_part2 = evaluate_indicators_part2(mock_data_part2)
print("=== 第9-13页经济指标评估体系 ===")
for key, value in eval_results_part2.items():
    print(f"{key}: {value}")
```

## 第四步：生成综合结论报告

```python
def generate_comprehensive_report(evaluation):
    """根据评估结果生成综合报告"""
    report = "【中国经济形势综合结论报告】\n"

    # 9. 固定资产投资
    inv_status = evaluation['fixed_investment']
    if inv_status == '健康':
        report += "- **固定资产投资**：投资总额保持稳健增长，显示政府和企业对未来经济前景有信心，基建和制造业投资是主要驱动力。\n"
    elif inv_status == '关注':
        report += "- **固定资产投资**：投资增速温和，民间投资活力有待提升，需关注政策对有效投资的引导作用。\n"
    else:
        report += "- **固定资产投资**：投资增速放缓甚至负增长，反映市场主体信心不足，经济面临下行压力，亟需加大稳投资力度。\n"

    # 10. 工业生产
    ind_status = evaluation['industrial_production']
    if ind_status == '健康':
        report += "- **工业生产**：工业增加值与企业利润双双增长，表明工业部门运行良好，产业链韧性较强，产能利用率高。\n"
    elif ind_status == '关注':
        report += "- **工业生产**：工业生产恢复缓慢，企业利润承压，可能受成本上升或需求疲软影响，需关注上游原材料价格波动。\n"
    else:
        report += "- **工业生产**：工业增加值和利润双降，预示经济景气度下行，企业经营困难加剧，需警惕产能过剩和债务风险。\n"

    # 11. 就业和收入
    emp_status = evaluation['employment_income']
    if emp_status == '健康':
        report += "- **就业与收入**：失业率稳定在较低水平，居民收入持续增长，消费能力有保障，社会稳定性强。\n"
    elif emp_status == '关注':
        report += "- **就业与收入**：青年群体失业率偏高，居民收入增长乏力，可能抑制消费意愿，需加强就业优先政策和收入分配改革。\n"
    else:
        report += "- **就业与收入**：失业率攀升，居民收入下降，民生压力增大，将严重制约内需扩张，是当前经济面临的最大挑战之一。\n"

    # 12. 通货膨胀和实际利率
    infl_status = evaluation['inflation_rates']
    if infl_status == '健康':
        report += "- **通胀与利率**：CPI温和可控，PPI企稳回升，实际利率为正，有利于储蓄和投资，宏观调控空间充足。\n"
    elif infl_status == '关注':
        report += "- **通胀与利率**：通胀压力或通缩风险并存，实际利率偏低，可能削弱储蓄意愿，需平衡稳增长与防风险的关系。\n"
    else:
        report += "- **通胀与利率**：出现明显通缩或恶性通胀，实际利率为负，财富效应受损，金融系统稳定性承压，需采取强力干预措施。\n"

    # 13. 对外贸易
    trade_status = evaluation['foreign_trade']
    if trade_status == '健康':
        report += "- **对外贸易**：进出口双双强劲增长，贸易顺差扩大，显示中国制造在全球市场的竞争力依然强大，外部需求稳定。\n"
    elif trade_status == '关注':
        report += "- **对外贸易**：外贸增长动能减弱，出口面临地缘政治和贸易壁垒挑战，需加快开拓新兴市场和推动产业升级。\n"
    else:
        report += "- **对外贸易**：外贸大幅萎缩，贸易逆差扩大，反映外部需求急剧恶化，对经济增长构成拖累，需寻求新的增长点。\n"

    report += "\n**总结**：当前中国经济呈现出结构性分化特征。一方面，部分领域如制造业、出口仍具韧性；另一方面，消费、房地产和民间投资等内生动力有待提振。建议政策制定者采取更加精准有力的措施，一方面稳住基本盘，另一方面激发新动能，以实现经济的高质量发展。"

    return report

# 生成并打印报告
final_report = generate_comprehensive_report(eval_results_part2)
print("\n" + final_report)
```
