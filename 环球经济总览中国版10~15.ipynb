```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
plt.rcParams['font.sans-serif'] = ['PingFang SC', 'Arial Unicode MS']  # Mac常用中文字体
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题
```

```python
def generate_mock_data_part3():
    """生成第14-18页所需的模拟数据"""
    data = {}

    # 图14: 房地产行业
    dates_house = pd.date_range(start='2015-01-01', end='2025-09-01', freq='M')
    new_house_price = np.random.uniform(-15, 15, len(dates_house))
    second_hand_house_price = np.random.uniform(-15, 15, len(dates_house))

    sales_area = np.random.uniform(-60, 120, len(dates_house))
    new_start_area = np.random.uniform(-60, 120, len(dates_house))
    completed_area = np.random.uniform(-60, 120, len(dates_house))

    domestic_loan = np.random.uniform(-30, 60, len(dates_house))
    total_funding = np.random.uniform(-30, 60, len(dates_house))

    data['real_estate'] = {
        'dates': dates_house,
        'new_house_price': new_house_price,
        'second_hand_house_price': second_hand_house_price,
        'sales_area': sales_area,
        'new_start_area': new_start_area,
        'completed_area': completed_area,
        'domestic_loan': domestic_loan,
        'total_funding': total_funding,
    }

    # 图15: 货币和信贷供应
    dates_money = pd.date_range(start='2015-01-01', end='2025-09-01', freq='M')
    m1 = np.random.uniform(-5, 30, len(dates_money))
    m2 = np.random.uniform(5, 25, len(dates_money))

    bank_loan = np.random.uniform(-10, 12, len(dates_money))
    gov_bond = np.random.uniform(-10, 12, len(dates_money))
    corp_direct_finance = np.random.uniform(-10, 12, len(dates_money))
    off_balance_sheet = np.random.uniform(-10, 12, len(dates_money))
    credit_impulse_total = np.random.uniform(-10, 12, len(dates_money))

    data['money_credit'] = {
        'dates': dates_money,
        'm1': m1,
        'm2': m2,
        'bank_loan': bank_loan,
        'gov_bond': gov_bond,
        'corp_direct_finance': corp_direct_finance,
        'off_balance_sheet': off_balance_sheet,
        'credit_impulse_total': credit_impulse_total,
    }

    # 图16: 货币政策工具
    dates_policy = pd.date_range(start='2015-01-01', end='2025-09-01', freq='M')
    reserve_ratio_large = np.linspace(20, 8, len(dates_policy))  # 模拟下降趋势
    reserve_ratio_small = np.linspace(18, 6, len(dates_policy))  # 模拟下降趋势

    loan_rate_1y = np.random.uniform(1, 6, len(dates_policy))
    lpr_1y = np.random.uniform(1, 6, len(dates_policy))
    lpr_5y = np.random.uniform(2, 6, len(dates_policy))
    reverse_repo_7d = np.random.uniform(1, 5, len(dates_policy))
    deposit_rate_1y = np.random.uniform(0.5, 3, len(dates_policy))

    data['monetary_policy'] = {
        'dates': dates_policy,
        'reserve_ratio_large': reserve_ratio_large,
        'reserve_ratio_small': reserve_ratio_small,
        'loan_rate_1y': loan_rate_1y,
        'lpr_1y': lpr_1y,
        'lpr_5y': lpr_5y,
        'reverse_repo_7d': reverse_repo_7d,
        'deposit_rate_1y': deposit_rate_1y,
    }

    # 图17: 财政政策
    dates_fiscal = pd.date_range(start='2015-01-01', end='2025-09-01', freq='M')
    fiscal_revenue = np.random.uniform(-30, 30, len(dates_fiscal))
    fiscal_expenditure = np.random.uniform(-30, 30, len(dates_fiscal))

    # 政府债券发行进度 (按月份)
    months = np.arange(1, 13)
    bond_2022 = np.cumsum(np.random.uniform(5, 15, 12))
    bond_2023 = np.cumsum(np.random.uniform(5, 15, 12))
    bond_2024 = np.cumsum(np.random.uniform(5, 15, 12))
    bond_2025 = np.cumsum(np.random.uniform(5, 15, 12))

    data['fiscal_policy'] = {
        'dates': dates_fiscal,
        'fiscal_revenue': fiscal_revenue,
        'fiscal_expenditure': fiscal_expenditure,
        'months': months,
        'bond_2022': bond_2022,
        'bond_2023': bond_2023,
        'bond_2024': bond_2024,
        'bond_2025': bond_2025,
    }

    # 图18: 汇率
    dates_exchange = pd.date_range(start='2016-01-04', end='2025-09-01', freq='M')
    usd_cny = np.random.uniform(6.0, 7.5, len(dates_exchange))
    cfets_index = np.random.uniform(90, 108, len(dates_exchange))

    # 添加事件标记
    event_dates = [pd.to_datetime('2018-04-01'), pd.to_datetime('2025-04-01')]  # 特朗普任期开始时间点

    data['exchange_rate'] = {
        'dates': dates_exchange,
        'usd_cny': usd_cny,
        'cfets_index': cfets_index,
        'event_dates': event_dates,
    }

    return data

# 生成模拟数据
mock_data_part3 = generate_mock_data_part3()
```

## 图10: 中国经济：房地产行业

```python
def plot_real_estate(data):
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))
    fig.suptitle('中国经济：房地产行业', fontsize=16)

    # 左上图：70个大中城市住宅价格
    ax1 = axes[0, 0]
    ax1.plot(data['dates'], data['new_house_price'], color='blue', label='新建商品住宅')
    ax1.plot(data['dates'], data['second_hand_house_price'], color='black', label='二手住宅')
    ax1.set_title('70个大中城市住宅价格\n同比', fontsize=12)
    ax1.set_ylabel('百分比 (%)')
    ax1.legend()

    # 右上图：房地产开发资金来源
    ax2 = axes[0, 1]
    ax2.plot(data['dates'], data['domestic_loan'], color='blue', label='国内贷款')
    ax2.plot(data['dates'], data['total_funding'], color='black', label='到位资金总量')
    ax2.set_title('房地产开发资金来源\n累计同比', fontsize=12)
    ax2.set_ylabel('百分比 (%)')
    ax2.legend()

    # 左下图：商品住宅市场面积指标
    ax3 = axes[1, 0]
    ax3.plot(data['dates'], data['sales_area'], color='blue', label='销售面积')
    ax3.plot(data['dates'], data['new_start_area'], color='green', label='新开工面积')
    ax3.plot(data['dates'], data['completed_area'], color='black', label='竣工面积')
    ax3.set_title('商品住宅市场面积指标\n累计同比', fontsize=12)
    ax3.set_ylabel('百分比 (%)')
    ax3.legend()

    # 右下图留空或添加其他内容
    ax4 = axes[1, 1]
    ax4.axis('off')

    plt.tight_layout()
    plt.show()

# 调用绘图函数
plot_real_estate(mock_data_part3['real_estate'])
```

## 图11: 中国经济：货币和信贷供应

```python
def plot_money_credit(data):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 7))

    # 左侧图：货币供应量
    ax1.plot(data['dates'], data['m1'], color='blue', label='M1')
    ax1.plot(data['dates'], data['m2'], color='black', label='M2')
    ax1.set_title('货币供应量*\n同比', fontsize=14)
    ax1.set_ylabel('百分比 (%)')
    ax1.legend()

    # 右侧图：信贷脉冲
    ax2.bar(data['dates'], data['bank_loan'], color='steelblue', label='银行贷款***', width=20)
    ax2.bar(data['dates'], data['gov_bond'], color='purple', label='政府债券', width=20, bottom=data['bank_loan'])
    ax2.bar(data['dates'], data['corp_direct_finance'], color='olive', label='企业直接融资****', width=20, bottom=data['bank_loan'] + data['gov_bond'])
    ax2.bar(data['dates'], data['off_balance_sheet'], color='orange', label='表外融资*****', width=20, bottom=data['bank_loan'] + data['gov_bond'] + data['corp_direct_finance'])
    ax2.plot(data['dates'], data['credit_impulse_total'], color='black', linewidth=2, label='信贷脉冲总量')

    ax2.set_title('信贷脉冲**\n百分点，新增社会融资和主要分项占名义GDP百分比的同比增减', fontsize=14)
    ax2.set_ylabel('百分点')
    ax2.legend(loc='upper left')

    plt.tight_layout()
    plt.show()

# 调用绘图函数
plot_money_credit(mock_data_part3['money_credit'])
```

## 图12: 中国经济：货币政策工具

```python
def plot_monetary_policy(data):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 7))

    # 左侧图：法定存款准备金率
    ax1.step(data['dates'], data['reserve_ratio_large'], where='post', color='black', label='大型银行')
    ax1.step(data['dates'], data['reserve_ratio_small'], where='post', color='blue', label='中小型银行')
    ax1.set_title('法定存款准备金率*', fontsize=14)
    ax1.set_ylabel('百分比 (%)')
    ax1.legend()

    # 右侧图：主要利率走势
    ax2.plot(data['dates'], data['loan_rate_1y'], color='black', label='1年期贷款基准**')
    ax2.plot(data['dates'], data['lpr_1y'], color='green', label='1年期LPR**')
    ax2.plot(data['dates'], data['lpr_5y'], color='olive', label='5年期LPR**')
    ax2.plot(data['dates'], data['reverse_repo_7d'], color='orange', label='7天银行间质押式回购 (30天移动平均)')
    ax2.plot(data['dates'], data['deposit_rate_1y'], color='purple', label='1年期存款***')
    ax2.set_title('主要利率走势\n年化利率', fontsize=14)
    ax2.set_ylabel('百分比 (%)')
    ax2.legend()

    plt.tight_layout()
    plt.show()

# 调用绘图函数
plot_monetary_policy(mock_data_part3['monetary_policy'])
```

## 图13 中国经济：财政政策

```python
def plot_fiscal_policy(data):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 7))

    # 左侧图：财政收入和支出
    ax1.plot(data['dates'], data['fiscal_revenue'], color='black', label='财政收入')
    ax1.plot(data['dates'], data['fiscal_expenditure'], color='blue', label='财政支出')
    ax1.set_title('财政收入和支出*\n同比，3个月滚动区间', fontsize=14)
    ax1.set_ylabel('百分比 (%)')
    ax1.legend()

    # 右侧图：政府债券发行进度
    ax2.plot(data['months'], data['bond_2022'], color='black', linestyle='--', label='2022年')
    ax2.plot(data['months'], data['bond_2023'], color='blue', linestyle='--', label='2023年')
    ax2.plot(data['months'], data['bond_2024'], color='green', linestyle='--', label='2024年')
    ax2.plot(data['months'], data['bond_2025'], color='purple', linestyle='--', label='2025年')
    ax2.set_title('政府债券**发行进度\n净发行量占全年发行限额比重', fontsize=14)
    ax2.set_xlabel('月份')
    ax2.set_ylabel('百分比 (%)')
    ax2.legend()

    plt.tight_layout()
    plt.show()

# 调用绘图函数
plot_fiscal_policy(mock_data_part3['fiscal_policy'])
```

## 图15  中国经济：汇率

```python
def plot_exchange_rate(data):
    fig, ax1 = plt.subplots(figsize=(14, 7))

    # 绘制人民币兑美元汇率 (左侧Y轴)
    ax1.plot(data['dates'], data['usd_cny'], color='black', label='人民币兑美元')
    ax1.set_title('人民币汇率\n美元/人民币，倒置刻度', fontsize=14)
    ax1.set_ylabel('美元/人民币', color='black')
    ax1.tick_params(axis='y', labelcolor='black')

    # 倒置Y轴
    ax1.invert_yaxis()

    # 添加箭头和注释
    ax1.annotate('人民币升值', xy=(data['dates'][0], 5.5), xytext=(data['dates'][0], 5.5),
                 arrowprops=dict(facecolor='black', shrink=0.05))
    ax1.annotate('人民币贬值', xy=(data['dates'][0], 7.5), xytext=(data['dates'][0], 7.5),
                 arrowprops=dict(facecolor='black', shrink=0.05))

    # 绘制CFETS人民币汇率指数 (右侧Y轴)
    ax2 = ax1.twinx()
    ax2.plot(data['dates'], data['cfets_index'], color='blue', label='CFETS人民币汇率指数* (右轴)')
    ax2.set_ylabel('指数', color='blue')
    ax2.tick_params(axis='y', labelcolor='blue')

    # 添加事件标记
    for event_date in data['event_dates']:
        ax1.axvline(x=event_date, color='red', linestyle='--', linewidth=1)
        ax1.text(event_date, 6.2, '特朗普第一任期内提高关税', ha='center', va='bottom', color='red', rotation=90)
        ax1.text(event_date, 6.0, '特朗普第二任期内提高关税', ha='center', va='top', color='red', rotation=90)

    # 添加图例
    lines1, labels1 = ax1.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

    plt.tight_layout()
    plt.show()

# 调用绘图函数
plot_exchange_rate(mock_data_part3['exchange_rate'])
```

## 第三步：构建指标评估体系

```python
def evaluate_indicators_part3(data):
    """评估第14-18页的关键经济指标"""
    evaluation = {}

    # 14. 房地产行业
    latest_new_house_price = data['real_estate']['new_house_price'][-1]
    latest_sales_area = data['real_estate']['sales_area'][-1]
    if latest_new_house_price > 0 and latest_sales_area > 0:
        evaluation['real_estate'] = '健康'
    elif latest_new_house_price > -5 or latest_sales_area > -10:
        evaluation['real_estate'] = '关注'
    else:
        evaluation['real_estate'] = '风险'

    # 15. 货币和信贷供应
    latest_m2 = data['money_credit']['m2'][-1]
    latest_credit_impulse = data['money_credit']['credit_impulse_total'][-1]
    if latest_m2 > 8 and latest_credit_impulse > 0:
        evaluation['money_credit'] = '健康'
    elif latest_m2 > 5 or latest_credit_impulse > -2:
        evaluation['money_credit'] = '关注'
    else:
        evaluation['money_credit'] = '风险'

    # 16. 货币政策工具
    latest_lpr_1y = data['monetary_policy']['lpr_1y'][-1]
    latest_reserve_ratio = data['monetary_policy']['reserve_ratio_large'][-1]
    if latest_lpr_1y < 4 and latest_reserve_ratio < 10:
        evaluation['monetary_policy'] = '健康'
    elif latest_lpr_1y < 5 or latest_reserve_ratio < 12:
        evaluation['monetary_policy'] = '关注'
    else:
        evaluation['monetary_policy'] = '风险'

    # 17. 财政政策
    latest_fiscal_deficit = data['fiscal_policy']['fiscal_expenditure'][-1] - data['fiscal_policy']['fiscal_revenue'][-1]
    if latest_fiscal_deficit > 5:
        evaluation['fiscal_policy'] = '健康'  # 积极财政政策
    elif latest_fiscal_deficit > 0:
        evaluation['fiscal_policy'] = '关注'
    else:
        evaluation['fiscal_policy'] = '风险'

    # 18. 汇率
    latest_usd_cny = data['exchange_rate']['usd_cny'][-1]
    latest_cfets_index = data['exchange_rate']['cfets_index'][-1]
    if 6.5 <= latest_usd_cny <= 7.0 and 95 <= latest_cfets_index <= 105:
        evaluation['exchange_rate'] = '健康'
    elif 6.0 <= latest_usd_cny <= 7.5 and 90 <= latest_cfets_index <= 110:
        evaluation['exchange_rate'] = '关注'
    else:
        evaluation['exchange_rate'] = '风险'

    return evaluation

# 执行评估
eval_results_part3 = evaluate_indicators_part3(mock_data_part3)
print("=== 第14-18页经济指标评估体系 ===")
for key, value in eval_results_part3.items():
    print(f"{key}: {value}")
```

## 第四步：生成综合结论报告

```python
def plot_exchange_rate_morgan_style(data):
    # 创建一个更大的画布，留出更多边距
    fig, ax1 = plt.subplots(figsize=(16, 8))

    # 设置背景为白色，无网格
    ax1.set_facecolor('white')
    ax1.grid(True, which='major', linestyle='-', linewidth=0.5, color='lightgray')

    # 绘制人民币兑美元汇率 (左侧Y轴)
    line1 = ax1.plot(data['dates'], data['usd_cny'], color='black', linewidth=2.5, label='人民币兑美元')
    ax1.set_title('人民币汇率', fontsize=18, fontweight='bold')
    ax1.set_ylabel('美元/人民币', fontsize=12, color='black')
    ax1.tick_params(axis='y', labelcolor='black', labelsize=11)

    # 倒置Y轴，让“升值”在上
    ax1.invert_yaxis()

    # 精确设置Y轴范围，突出变化
    ax1.set_ylim(7.6, 5.5)  # 手动设定范围，让波动看起来更剧烈

    # 添加箭头和注释
    # 人民币升值箭头
    ax1.annotate('', xy=(data['dates'][0], 5.8), xytext=(data['dates'][0], 5.6),
                 arrowprops=dict(facecolor='black', edgecolor='black', shrink=0.05, width=2, headwidth=8))
    ax1.text(data['dates'][0], 5.55, '人民币升值', ha='center', va='bottom', fontsize=10, fontweight='bold', color='black')

    # 人民币贬值箭头
    ax1.annotate('', xy=(data['dates'][0], 7.4), xytext=(data['dates'][0], 7.2),
                 arrowprops=dict(facecolor='black', edgecolor='black', shrink=0.05, width=2, headwidth=8))
    ax1.text(data['dates'][0], 7.15, '人民币贬值', ha='center', va='top', fontsize=10, fontweight='bold', color='black')

    # 绘制CFETS人民币汇率指数 (右侧Y轴)
    ax2 = ax1.twinx()
    line2 = ax2.plot(data['dates'], data['cfets_index'], color='steelblue', linewidth=2.5, label='CFETS人民币汇率指数* (右轴)')
    ax2.set_ylabel('指数', fontsize=12, color='steelblue')
    ax2.tick_params(axis='y', labelcolor='steelblue', labelsize=11)

    # 精确设置右侧Y轴范围
    ax2.set_ylim(88, 109)

    # 添加事件标记
    for event_date in data['event_dates']:
        ax1.axvline(x=event_date, color='red', linestyle='--', linewidth=1.5, zorder=5)
        # 在线上方添加注释
        ax1.text(event_date, 6.2, '特朗普第一任期内提高关税', ha='center', va='bottom',
                 fontsize=9, fontweight='bold', color='red', rotation=90, bbox=dict(boxstyle="round,pad=0.3", facecolor='white', edgecolor='red', alpha=0.8))
        # 在线下方添加注释
        ax1.text(event_date, 6.0, '特朗普第二任期内提高关税', ha='center', va='top',
                 fontsize=9, fontweight='bold', color='red', rotation=90, bbox=dict(boxstyle="round,pad=0.3", facecolor='white', edgecolor='red', alpha=0.8))

    # 添加图例，位置放在左下角，避免遮挡图形
    lines1, labels1 = ax1.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax1.legend(lines1 + lines2, labels1 + labels2, loc='lower left', fontsize=10, frameon=True, fancybox=True)

    # 移除顶部和右侧的边框
    ax1.spines['top'].set_visible(False)
    ax1.spines['right'].set_visible(False)
    ax2.spines['top'].set_visible(False)
    ax2.spines['right'].set_visible(False)

    # 调整布局，确保标题和标签不被裁剪
    plt.tight_layout(rect=[0, 0, 1, 0.95])

    # 添加摩根的页码和标识
    fig.text(0.95, 0.02, '中国版 | 18', ha='right', va='bottom', fontsize=10, fontweight='bold')
    fig.text(0.85, 0.02, '摩根资产管理', ha='right', va='bottom', fontsize=14, fontweight='bold')

    plt.show()

# 调用新函数
plot_exchange_rate_morgan_style(mock_data_part3['exchange_rate'])
```
